```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('diabetes_clean.csv')
```

a) set a seed (to ensure work reproducibility) and take a random sample of 25 observations and
find the mean Glucose and highest Glucose values of this sample and compare these statistics
with the population statistics of the same variable. You should use charts for this comparison.

```python
sample_data = data.sample(n=25)
glucose_mean = data['Glucose'].mean()
glucose_max = data['Glucose'].max()
sample_mean_glucose = sample_data['Glucose'].mean()
sample_max_glucose = sample_data['Glucose'].max()


colors = ['blue', 'green', 'orange', 'red']
plt.figure(figsize=(10, 5))
plt.bar(['Population Mean', 'Population Max', 'Sample Mean', 'Sample Max'],
        [glucose_mean, glucose_max, sample_mean_glucose, sample_max_glucose],
        color=colors)  # Specify colors
plt.title('Comparison of Glucose Statistics (Population vs Sample)')
plt.ylabel('Glucose')
plt.show()
```

b) Find the 98th percentile of BMI of your sample and the population and compare the results
using charts.

```python
population_98thbmi = np.percentile(data['BMI'], 98)
sample_98thbmi = np.percentile(sample_data['BMI'], 98)
colors = ['blue', 'green']

plt.figure(figsize=(8, 5))
plt.bar(['Population 98th Percentile', 'Sample 98th Percentile'],
        [population_98thbmi, sample_98thbmi],
        color=colors)  # Specify colors
plt.title('Comparison of 98th Percentile of BMI (Population vs Sample)')
plt.ylabel('BMI')
plt.show()
```

c) Using bootstrap (replace= True), create 500 samples (of 150 observation each) from the
population and find the average mean, standard deviation and percentile for BloodPressure and
compare this with these statistics from the population for the same variable. Again, you should
create charts for this comparison. Report on your findings

```python
population_mean_bp = data['BloodPressure'].mean()
population_std_bp = data['BloodPressure'].std()
population_percentile_bp_25 = np.percentile(data['BloodPressure'], 25)  
population_percentile_bp_50 = np.percentile(data['BloodPressure'], 50)
population_percentile_bp_75 = np.percentile(data['BloodPressure'], 75)  
population_percentile_bp_95 = np.percentile(data['BloodPressure'], 95)  


n_bootstrap = 500
bootstrap_sample_size = 150

bootstrap_means = []
bootstrap_stds = []
bootstrap_percentiles_25 = []
bootstrap_percentiles_50 = []
bootstrap_percentiles_75 = []
bootstrap_percentiles_95 = []

for _ in range(n_bootstrap):
    bootstrap_sample = np.random.choice(data['BloodPressure'], size=bootstrap_sample_size, replace=True)
    bootstrap_means.append(np.mean(bootstrap_sample))
    bootstrap_stds.append(np.std(bootstrap_sample))
    bootstrap_percentiles_25.append(np.percentile(bootstrap_sample, 25))
    bootstrap_percentiles_50.append(np.percentile(bootstrap_sample, 50))
    bootstrap_percentiles_75.append(np.percentile(bootstrap_sample, 75))
    bootstrap_percentiles_95.append(np.percentile(bootstrap_sample, 95))


plt.figure(figsize=(20, 5))

plt.subplot(1, 5, 1)
plt.hist(bootstrap_means, bins=30, alpha=0.5, color='blue', label='Bootstrap Sample Means')
plt.axvline(x=population_mean_bp, color='red', linestyle='--', label='Population Mean')
plt.title('Comparison of Mean BloodPressure')
plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.legend()


plt.subplot(1, 5, 2)
plt.hist(bootstrap_stds, bins=30, alpha=0.5, color='green', label='Bootstrap Sample Std')
plt.axvline(x=population_std_bp, color='red', linestyle='--', label='Population Std')
plt.title('Comparison of Standard Deviation of BloodPressure')
plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.legend()


plt.subplot(1, 5, 3)
plt.hist(bootstrap_percentiles_25, bins=30, alpha=0.5, color='purple', label='Bootstrap Sample 25th Percentiles')
plt.axvline(x=population_percentile_bp_25, color='red', linestyle='--', label='Population 25th Percentile')
plt.title('Comparison of 25th Percentile of BloodPressure')
plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 5, 4)
plt.hist(bootstrap_percentiles_50, bins=30, alpha=0.5, color='orange', label='Bootstrap Sample 50th Percentiles')
plt.axvline(x=population_percentile_bp_50, color='red', linestyle='--', label='Population 50th Percentile')
plt.title('Comparison of 50th Percentile of BloodPressure')
plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 5, 5)
plt.hist(bootstrap_percentiles_75, bins=30, alpha=0.5, color='cyan', label='Bootstrap Sample 75th Percentiles')
plt.axvline(x=population_percentile_bp_75, color='red', linestyle='--', label='Population 75th Percentile')
plt.title('Comparison of 75th Percentile of BloodPressure')
plt.xlabel('BloodPressure')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
